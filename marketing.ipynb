### Deploying K-Means Algo

```python

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
scaler = StandardScaler()
```

```python
scaled_df= scaler.fit_transform(df)
```

```python
scaled_df
```

Search for the best number of clusters

```python
ssd = []

for k in range(2,30):
    
    model = KMeans(n_clusters=k)
    
    
    model.fit(scaled_df)
    
    #Sum of squared distances of samples to their closest cluster center.
    ssd.append(model.inertia_)
```

```python
plt.figure(figsize=(12,7))
plt.plot(range(2,30),ssd,'o--')
plt.xlabel("K Value")
plt.ylabel(" Sum of Squared Distances")
```

```python
plt.figure(figsize=(12,7))
pd.Series(ssd).diff().plot(kind='bar')
```

elbow curve and barchart show that 8 clusters are okay 

#### Deploying the K-means with 8 clusters

```python
kmeans = KMeans(n_clusters=8)
kmeans.fit(scaled_df)
labels=kmeans.labels_
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

visualizing the cluster centers as a data frame

```python
cluster_centers = pd.DataFrame(data = kmeans.cluster_centers_, columns = [df.columns])
cluster_centers 
```

converting the cluster centers to their orgianl scale (inverse transform)

```python
cluster_centers = scaler.inverse_transform(cluster_centers)
cluster_centers = pd.DataFrame(data = cluster_centers, columns = [df.columns])
cluster_centers.head()
```

```python
df_cluster = pd.concat([df, pd.DataFrame({'cluster':labels})], axis = 1)
df_cluster.head()
```

### Visualising the clusters and their features

```python
for i in df.columns:
  plt.figure(figsize = (35, 5))
  for j in range(8):
    plt.subplot(1,8,j+1)
    cluster = df_cluster[df_cluster['cluster'] == j]
    cluster[i].hist(bins = 20)
    plt.title('{}    \nCluster {} '.format(i,j))
  
  plt.show()
```
